Найти медиану, моду, математическое ожидание, стандартное отклонение,  дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
# Проверим наши данные на выбросы
print("Выбросов =",df['Freedom'].isnull().sum())
# Заменим несуществующие значения на ноль
df=df.fillna(0)
print("математическое ожидание =",statistics.mean(df['Freedom']))
print("медиана =",statistics.median(df['Freedom']))
print("минимум =",min(df['Freedom']))
print("максимум =",max(df['Freedom']))
print("размах =",max(df['Freedom'])-min(df['Freedom']))
print("стандартное отклонение =",statistics.pstdev(df['Freedom']))
print("дисперсия =",statistics.pvariance(df['Freedom']))
print('Значение моды: ', statistics.mode(df['Freedom']))
# Исходя из ошибки - мода не может быть выделена, все значения встречаются единожды
```

Посмотреть корреляцию между Freedom и positive affect (графически и математически). Графически - plt.scatter, математически - посчитать коэффициент Пирсона. О чем говорит знак коэффициента?

```python
Freedom = df['Freedom']
positive_affect = df['Positive affect']
plt.xlabel('выборка Freedom')
plt.ylabel('выборка positive_affect')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(Freedom, positive_affect, marker='.')
```

```python
from scipy.stats import pearsonr
corr, p_value = pearsonr(Freedom, positive_affect)
print("Коэффициент корреляции пирсона =", corr)
# Коэффициент имеет знак плюс, что говорит о прямой зависимости двух величин
```

Нормализовать данные колонки Freedom (централизация + масштабирование)

```python
Freedom_norm=Freedom - statistics.mean(Freedom)
print(statistics.mean(Freedom_norm))

from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))
Freedom_scaled = min_max_scaler.fit_transform(np.array(Freedom_norm).reshape(-1, 1))
print(min(Freedom_scaled))
print(max(Freedom_scaled))
```
